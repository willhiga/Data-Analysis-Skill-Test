```
fig = go.Figure()

fig.add_trace(go.Bar(
    x=["Milho", "Açúcar", "Trigo", "Grãos de soja", "Óleo de soja", "Carne de soja"],
    y=[crnt, sgrt, wht, sbt, sot, smt]
))

fig.update_layout(
    autosize=False,
    width=750,
    height=750,
    yaxis=dict(
        title_text="Total de exportações nos últimos 5 anos (USD)",
        tickmode="array",
        titlefont=dict(size=20),
    )
)

fig.update_yaxes(automargin=True)

fig.show()
```

**Resposta**

Pelo gráfico acima percebe-se que os principais produtos exportados nos últimos 5 anos pelo Brasil são: Grão de soja, açúcar e carne de soja.

**Resposta**

Pelo gráfico acima é possível observar que nos últimos 5 anos as principais rotas de exportação de milho foram do tipo marítima e terrestre. Tal predominância pela via marítima se deve ao fato de a maior parte dos mercados importadores estarem situados em outros continentes, daí a escolha pela rota marítima. A escolha do tipo de rota adequada é de vital importância devido a sua influência no custo final de venda.
